```python
from sklearn.manifold import TSNE
import pandas
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
dataframe = pandas.read_csv('heart.csv',usecols = range(0,13), engine='python')
X = dataframe.values
dataframe = pandas.read_csv('heart.csv',usecols = [13], engine='python')
dataset = dataframe.values
Y = []
for i in dataset:
    Y.append(i[0])
```

```python
tsne = TSNE(n_components=3, learning_rate=40)
X_tsne = tsne.fit_transform(X)
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], X_tsne[:, 2], c=Y)
plt.figure(figsize=(175,100))
# fig = plt.figure()
# ax = fig.add_subplot(111, projection='3d')

# ax.scatter
# print X_tsne
# plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=Y, cmap=plt.cm.cool)
# print Y
plt.show()
```
